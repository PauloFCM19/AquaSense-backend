```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

Lendo os dados de medição

```python
data = pd.read_csv('/content/historicoDeMedidasMaio.csv', sep=';', encoding='latin1')
```

Removendo as colunas que não serão utilizadas no momento, para obtenção da média mensal


```python
dataFilter = data.drop(columns=['Amostra', 'Altura(CM)', 'Consumo Mensal', 'Dias'])
dataFilterMensal = data.drop(columns=['Amostra', 'Altura(CM)', 'Data', 'Nivel[%]', 'Volume[L]'])

```

```python
dataFilter.plot(x="Data", y="Nivel[%]", kind='bar' )
plt.title('')
plt.xlabel('Maio')
plt.ylabel('')
plt.yticks(rotation=0)

plt.show()
```

Lendo o arquivo da medida atual e fazendo o tratamento para converte-lo em gráfico.

```python
dataAtual = pd.read_csv('/content/MedidaAtual.csv', sep=';', encoding='latin1')
plot = dataAtual.plot(x="Data", y="Nivel[%]", kind='bar' )
for i in plot.patches:
   plot.annotate(str(round(i.get_height(), 1)), (i.get_x() * 0.100, i.get_height() * 1.010))
plt.title('Medida Atual 65%')
plt.xlabel('')
plt.xticks(rotation=0)
plt.ylabel('')
plt.ylim(0, 100)
plt.yticks(np.arange(0, 101, 5))
plt.show()
```

```python
dataFilterMensal['Media Dia'] = dataFilterMensal['Consumo Mensal'] / dataFilterMensal['Dias']
dataFilterDia = dataFilterMensal.drop(index=range(1, 25))
plot = dataFilterDia.plot(x="Dias", y="Media Dia", kind='bar' )
for i in plot.patches:
   plot.annotate(str(round(i.get_height(), 1)), (i.get_x() * 0.100, i.get_height() * 1.010))
plt.title('Média Diária')
plt.xlabel('Quantidade de dias')
plt.ylabel('')
plt.ylim(0, 100)
plt.yticks(np.arange(0, 101, 5))
plt.xticks(rotation=0)







```
